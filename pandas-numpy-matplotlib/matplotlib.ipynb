```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize='small')
```

```python
ax.set_title('my plot')
ax.set_xlabel('count')
```

```python
props={'title':'my first plot','xlabel':'stages'}
ax.set(**props)
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(),'b',label='one')
ax.plot(np.random.randn(1000).cumsum(),'r--',label='two')
ax.plot(np.random.randn(1000).cumsum(),'g-',label='three')
plt.legend(loc='best')
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
rect=plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3)
circle=plt.Circle((0.7,0.2),0.15,color='b',alpha=0.3)
pgon=plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color='g',alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circle)
ax.add_patch(pgon)
```
