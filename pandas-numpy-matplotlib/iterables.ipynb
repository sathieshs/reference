```python
## Example iterateor class

class ExampleIterator:
    
    def __init__(self):
        self.data=[1,2,3,4,5]
        self.index=0
        
    def __iter__(self):
        return self
    
    def __next__(self):
        currval=self.index
        self.index=self.index+1
        if(self.index==len(self.data)):
           raise stopIteration()
        return self.data[currval]
        
```
