## discreatization

```python
ages=[20,22,25,27,21,23,37,31,61,45,41,32]
```

```python
bins=[18,25,35,60,100]
import pandas as pd
cats=pd.cut(ages,bins)
```

```python
cats
```

```python
cats.codes
```

```python
cats.categories
```

```python
pd.value_counts(cats)
```

```python
groupnames=['youth','youngAdult','middleAged','senior']
cats2=pd.cut(ages,bins,labels=groupnames)
```

```python
cats2
```

```python
pd.value_counts(cats2)
```

```python
import numpy as np
data=pd.Series(np.random.randn(1000))
cat3=pd.cut(data,10,precision=2)
```

```python
pd.value_counts(cat3).
```

```python
cat4=pd.qcut(data,4)
pd.value_counts(cat4)
```
